```python
import pandas as pd
```

```python
dt = pd.read_csv('potato2019.csv')
dt
```

```python
feature = ['state','district','market','commodity','variety','arrival_date','min_price','max_price']
x = dt[feature]
y = dt.modal_price
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
dt.state_encoded=le.fit_transform(dt.state)
print(dt.state_encoded)
dt.district_encoded=le.fit_transform(dt.district)
print(dt.district_encoded)
dt.market_encoded=le.fit_transform(dt.market)
print(dt.market_encoded)
dt.variety_encoded=le.fit_transform(dt.variety)
print(dt.variety_encoded)
dt.arrival_date_encoded=le.fit_transform(dt.arrival_date)
print(dt.arrival_date_encoded)
```

```python
x=list(zip(dt.state_encoded,dt.district_encoded,dt.market_encoded,dt.variety_encoded,dt.arrival_date_encoded,dt.min_price,dt.max_price))
print(x)
```

```python
from matplotlib import pyplot as plt 
import seaborn as sns
# Calculate correlations
corr = dt.corr()
 
# Heatmap
sns.heatmap(corr)
```

```python
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=100) 
# 80% training and 20% test
```

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
reg = LinearRegression()
reg.fit(x_train,y_train)
accuracy = reg.score(x_test,y_test)
print(accuracy*100,'%')
```
